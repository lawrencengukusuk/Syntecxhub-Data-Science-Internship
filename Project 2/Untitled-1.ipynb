```python
#Filter Using .isin()

#Orders from North and South regions:

df[df["Region"].isin(["North", "South"])]

# Filter by Sales Representative
df[df["Sales_Rep"] == "Bob"]
```

```python
#Creating a Subset Called online_electronics
#Suppose I want to analyze online electronics sales:

online_electronics = df[
    (df["Sales_Channel"] == "Online") &
    (df["Product_Category"] == "Electronics")
]

online_electronics.head()
```
